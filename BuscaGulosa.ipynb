```python
class Vertice:
    def __init__(self, rotulo, distancia_objetivo):
        self.rotulo = rotulo
        self.visitado = False
        self.distancia_objetivo = distancia_objetivo
        self.adjacente = []

    def adiciona_adjacente(self, adjacente):
        self.adjacente.append(adjacente)

    def mostra_adjacente(self):
        for adj in self.adjacente:
            print(adj.vertice.rotulo, adj.custo)
```

```python
class VetorOrdenado:
    def __init__(self, capacidade: int):
        self.capacidade = capacidade
        self.ultima_posicao = -1
        self.valores = np.empty(self.capacidade, dtype=object)

    def imprime(self):
        if self.ultima_posicao == -1:
            print("Vetor vazio")
            return
        for i in range(self.ultima_posicao + 1):
            print(i, "-", self.valores[i], "-", self.valores[i].distancia_objetivo)

    def insere(self, vertice):
        if self.ultima_posicao == self.capacidade - 1:
            print("Capacidade máxima atingida")
            return

        if self.ultima_posicao == -1:
            posicao = 0
        else:
            posicao = self.ultima_posicao + 1
            for i in range(self.ultima_posicao + 1):
                if self.valores[i].distancia_objetivo > vertice.distancia_objetivo:
                    posicao = i
                    break

        for i in range(self.ultima_posicao, posicao - 1, -1):
            self.valores[i + 1] = self.valores[i]

        self.valores[posicao] = vertice
        self.ultima_posicao += 1

    def get_primeiro(self):
        if self.ultima_posicao == -1:
            return None
        return self.valores[0]
```

```python
class Gulosa:
    def __init__(self, objetivo):
        self.objetivo = objetivo
        self.encontrado = False
        self.fronteira = VetorOrdenado(20)
        
    def buscar(self, atual):
        print(f"\nPosição atual: {atual.rotulo}")
        
        opcoes = []
        for adjacente in atual.adjacente:
            if not adjacente.vertice.visitado:
                opcoes.append(f"{adjacente.vertice.rotulo} (custo: {adjacente.custo}, dist: {adjacente.vertice.distancia_objetivo})")
        
        if opcoes:
            print(f"Opções: {', '.join(opcoes)}")
        else:
            print("Opções: nenhuma opção disponível")
        
        atual.visitado = True
        
        if atual == self.objetivo:
            self.encontrado = True
            print("\nObjetivo encontrado!")
            return
            
        for adjacente in atual.adjacente:
            if not adjacente.vertice.visitado:
                self.fronteira.insere(adjacente.vertice)
        
        if not self.encontrado:
            proximo = self.fronteira.get_primeiro()
            if proximo is not None:
                print("-" * 30)
                self.buscar(proximo)
            else:
                print("\nCaminho não encontrado!")
```

```python
def buscar(self, atual):
    print(" - "*30)
    print(f"Atual: {atual.rotulo}")
    atual.visitado = True

    if atual == self.objectivo:
        self.encontrado = True
    else:
        vetor_ordenado = VetorOrdenado(len(atual.adjacente))
        for adjacente in atual.adjacente:
            if adjacente.vertice.visitado == False:
                adjacente.vertice.visitado = True
                vetor_ordenado.insere(adjacente)
        vetor_ordenado.imprime()

        if vetor_ordenado.valores[0] != None:
            self.buscar(vetor_ordenado.valores[0])
```
