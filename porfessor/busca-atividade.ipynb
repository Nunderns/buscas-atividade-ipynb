```python
class Adjacente:
    def __init__(self, vertice, custo):
        self.vertice = vertice
        self.custo = custo


class Vertice:
    def __init__(self, nome, estimativa):
        self.nome = nome
        self.estimativa = estimativa
        self.adjacentes = []

    def adiciona_adjacente(self, adjacente):
        self.adjacentes.append(adjacente)

    def __repr__(self):
        return self.nome
```

```python
def heuristica(vertice_atual, entregas_restantes):
    """Heurística simples: menor estimativa entre os vértices restantes."""
    if not entregas_restantes:
        return 0
    return min(v.estimativa for v in entregas_restantes)


def reconstruir_caminho(anteriores, estado_final):
    """Reconstrói a rota completa a partir do dicionário de predecessores."""
    caminho = []
    atual = estado_final
    while atual:
        vert, restantes = atual
        caminho.append(vert.nome)
        atual = anteriores.get(atual)
    caminho.reverse()
    return caminho
```

```python
import heapq

def a_star_multientrega(grafo, inicio, entregas):
    estado_inicial = (inicio, tuple(entregas))
    fronteira = []
    heapq.heappush(fronteira, (0, estado_inicial))
    custos = {estado_inicial: 0}
    anteriores = {estado_inicial: None}

    while fronteira:
        _, (atual, restantes) = heapq.heappop(fronteira)
        if not restantes:
            caminho = reconstruir_caminho(anteriores, (atual, restantes))
            return caminho, custos[(atual, restantes)]

        for adj in atual.adjacentes:
            proximo = adj.vertice
            novo_restante = tuple(r for r in restantes if r != proximo)
            novo_estado = (proximo, novo_restante)
            novo_custo = custos[(atual, restantes)] + adj.custo

            if novo_estado not in custos or novo_custo < custos[novo_estado]:
                custos[novo_estado] = novo_custo
                h = heuristica(proximo, novo_restante)
                prioridade = novo_custo + h
                heapq.heappush(fronteira, (prioridade, novo_estado))
                anteriores[novo_estado] = (atual, restantes)

    return None, float("inf")
```
